##### Statistical Analysis

```python
from scipy.stats import chisquare

def compute_categorical_data(df):
  results = {}
  for col in df.select_dtypes(include=['object']).columns:
    observed_frequencies = df[col].value_counts()
    
    # For a uniform distribution, each category is expected to have the same frequency
    expected_frequency = [len(categorical_data) / len(observed_frequencies)] * len(observed_frequencies)
    
    # Perform the chi-square test
    chi_square_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequency)
    
    results[col] = {
        'Chi-square statistic': chi_square_stat,
        'P-value': p_value
    }
    
  return pd.DataFrame(results);
  

compute_categorical_data(categorical_data)
```

Because all the p-values are essentially rounded to zero, it means that for each variable, the distribution of categories in each variable is very unlikely to be uniform, and there are certain categories that are significantly more or less common than others.

This therefore presents a strong evidence against the null hypothesis for each variable


Examples of null hypotheses:
- **GEO_UNIT:** The distribution of categories within GEO_UNIT fits the expected distribution 
- **Climate_Type:** The distribution of categories within Climate_Type fits the expected distribution.
- **DES_ClimateType:** The distribution of categories within DES_ClimateType fits the expected distribution.

## Outlier Treatment

```python
import numpy as np

def treat_outliers_with_iqr(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        
        upper_bound = Q3 + 1.5 * IQR
        
        # Replace outliers with the nearest non-outlier value
        data[col] = np.where(data[col] < lower_bound, lower_bound,
                             np.where(data[col] > upper_bound, upper_bound, data[col]))
    
    return data
```
